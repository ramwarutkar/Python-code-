```python
""" Write a Python program to get the least common multiple (LCM) of two positive integers. """
def gcd(a,b): 
    if a==0: 
        return b 
    return gcd(b % a, a) 
 
def lcm(a,b): 
    return (a*b) / gcd(a,b) 
a = 15 
b = 20
print('LCM of', a, 'and', b, 'is', lcm(a, b)) 
```

```python
""" Write a program to calculate area of various shapes like circle, rectangle and square using function. """


def area_shape(input_shape):
    if input_shape=='Circle':
        input_c_radius=int(input("Enter the Radius - "))
        return 3.14*(input_c_radius**2)
    elif input_shape=='Rectangle':
        input_r_width=int(input("Enter the Width of rectangle - "))
        input_r_height=int(input("Enter the Height of rectangle - "))
        return input_r_width*input_r_height
    elif input_shape=='Square':
        input_s_side=int(input("Enter the Side of Square - "))
        return input_s_side**2
    elif input_shape=='Triangle':
        input_t_height=int(input("Enter the Height of Triangle - "))
        input_t_base=int(input("Enter the Base of Triangle - "))
        return 0.5*input_t_height*input_t_base
print("\n1. Circle \n2. Rectangle\n3. Square\n4. Triangle")
input_shape=input("Enter the Shape - ")
print("Area is - ",area_shape(input_shape))
```

```python
""" Write a program to calculate Simple Interest using function. Take the values from the user for P, N, and R. 
Calculating rate for per year 
Simple Interst - Pinciple Amount*Rate*(Month/12) """

def simple_interst(input_principle,input_rate,input_time_month):
    return (input_principle*(input_rate)*(input_time_month/12))/100

input_principle=int(input("Enter the Principle Amount - "))
input_rate=float(input("Enter the rate in percent - "))
input_time_month=int(input("Enter the time in Months - "))
print("Simple Intrest - ",simple_interst(input_principle,input_rate,input_time_month))
```

```python
""" Write a Python function that accepts student names and marks (store in list). Calculate the total marks
and average based on the input. Display student names and marks, total and average. """

infinity=1

list_name=[]
list_marks=[]
input_name=''
input_marks=0
total_marks=0
stu_average=0
def stu_marks_avg(list_name,list_marks):
    input_name=input("Enter the Name of student - ")
    list_name.append(input_name)
    input_marks=int(input("Enter the marks of student - "))
    list_marks.append(input_marks)
while infinity==1:
    input_action=input("Enter Y/N - ")
    if input_action=='N':
        break
    else:
        stu_marks_avg(list_name,list_marks)
for i in range(len(list_marks)):
    total_marks=total_marks+list_marks[i]
stu_average=total_marks/len(list_marks)
for j in range(len(list_marks)):
    print("Student - ",list_name[j]," Marks - ",list_marks[j])
print("Average Marks is - ",stu_average)
        
        
    
    
    
```

```python
""" Write a program to count the number of each vowel in a string using dictionary and list comprehension. """
vowel=['a','e','i','o','u']
string='aeiou'
dic={}
v_count=0
for i in range(len(string)):
    dic.update({i:string[i]})
print(dic)
for i in range(len(dic)):
    if dic.get(i) in vowel:
        v_count=v_count+1
print(v_count)
```

```python
""" Write a program to remove all the punctuations from a string and check each character of the string
using for loop, If the character is a punctuation, assign an empty string to it. """

str="Pyth@on@! asdf$#dfs^"
output_string=''
for i in range(len(str)):
    if str[i].isalpha():
        output_string=output_string+str[i]
    else:
        print("Alpha")
print(output_string)
        
```
